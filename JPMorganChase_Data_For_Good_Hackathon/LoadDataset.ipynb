```python
def count(table_name, table_data):
    count_field = {}
    for col in table_data.columns:
        count_field[col] = {}
    
    
        
    for i in range(len(table_data)):
        x = table_data.iloc[i]
        for i in x.index:
            if x[i] in count_field[i]:
                count_field[i][x[i]] += 1
            else:
                count_field[i][x[i]]  = 1
            
        
        
    return count_field

def rank_dictionary(dic):
    dic_view = [ (v,k) for k,v in dic.items() ]
    dic_view.sort(reverse=True) # natively sort tuples by first element
    for v,k in dic_view:
        print("%s: %d" % (k,v))
```

```python
counts_review = count("results-20251009-150104", dataset_review["results-20251009-150104"])
```
