```python
success = ["COMPLETED", "ACTIVE"]
df_all["status_flag"] = df_all["binome_statut"].apply(
    lambda x: 1 if x in success else 0
)
```

```python
df_all = df_all.drop(columns=["Unnamed: 14","Unnamed: 13","mentee_id","mentor_id"])
```

```python
df_all = df_all.dropna(subset=["workfield","current_role"], how="all")
```

```python
df_all.loc[df_all["binome_cancellation_reason"].isna() & (df_all["status_flag"] == 0), "binome_cancellation_reason"] = "Not specified (or Not provided)"
df_all.loc[df_all["binome_cancellation_reason"].isna() & (df_all["status_flag"] == 1), "binome_cancellation_reason"] = "Not applicable"
```

```python
df_all["desired_exchange_frequency"] = df_all["desired_exchange_frequency"].fillna("Not specified (or Not provided)")
```
